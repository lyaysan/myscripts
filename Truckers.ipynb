```python
import pandas as pd
import numpy as np
import dateutil.parser
```

```python
""" load_postings.csv """
data_post = pd.read_csv("C:\\Users\\Ляйсан\\Desktop\\Trcukerpath test assignment\\load_postings.csv")
```

```python
data_post = data_post.drop('Unnamed: 0', axis = 1)
```

```python
""" search available orders - delete available at future (we need available during the week) """

data_post = data_post[data_post.from_date_local < '2017-01-30 00:00:00+00:00']

""" delete duplicates - orders with simular creating time (YYYYMMDD HHMM), user id, company, from-to parameters, equipment ect"""

data_post['created_YMD'] = data_post['created_at'].apply(lambda x: dateutil.parser.parse(x).strftime("%Y-%m-%d %H-%M")).head()

data_post.drop_duplicates(['created_YMD', 'user_id', 'company', 'from_date_local', 'from_state', 'from_address', 'to_state', 'to_address', 'equipment', 'weight', 'load_size', 'distance', 'comments'])
```

```python
""" truckloads_searches.csv """

data_search = pd.read_csv("C:\\Users\\Ляйсан\\Desktop\\Trcukerpath test assignment\\truckloads_searches.csv")
data_search = data_search.drop('Unnamed: 0', axis = 1)
```

```python
""" search available trackers - delete available at future (we need available during the week) """

data_search = data_search[data_search['Pickup Date'] < '2017-01-30 00:00:00+00:00']

""" we need only users (not searchs) """

data_search = data_search.drop_duplicates('distinct_id')
```

```python
""" group data by states """

x = data_post.groupby('from_state')['id'].apply(np.size)
y = data_search.groupby('Pickup State')['distinct_id'].apply(np.size)

ratio = (x/y).dropna()
```
